```python
import pywebcopy
import pywebcopy.configs
print(pywebcopy.__version__.__version__)

conf = pywebcopy.configs.default_config
print(conf)

conf = pywebcopy.configs.get_config(
    
    project_url=project_url, 
    project_folder=project_folder, 
    project_name=project_name,
    
    #overwrite=False,
    
    bypass_robots=False,
    debug=False,
    delay=None,
    threaded=None,
)
print(conf)

crawler = pywebcopy.Crawler()
crawler.crawl()
```
